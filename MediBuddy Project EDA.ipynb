## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
print("\nPrice Data Types:")
print(price_data.dtypes)

print("\nPersonal Data Types:")
print(personal_data.dtypes)

# Convert categorical variables to string (if necessary)
if 'gender' in personal_data.columns:
    personal_data['gender'] = personal_data['gender'].astype(str)
if 'smoker' in personal_data.columns:
    personal_data['smoker'] = personal_data['smoker'].astype(str)
if 'region' in personal_data.columns:
    personal_data['region'] = personal_data['region'].astype(str)

````

````python
# Check for missing values in both datasets
print("\nMissing Values in Price Data:")
print(price_data.isnull().sum())

print("\nMissing Values in Personal Data:")
print(personal_data.isnull().sum())
````

````python
# Remove duplicate rows
price_data.drop_duplicates(inplace=True)
personal_data.drop_duplicates(inplace=True)
````

````python
price_data.reset_index(drop=True, inplace=True)
personal_data.reset_index(drop=True, inplace=True)

# Concatenate both datasets along columns (axis=1)
merged_data = pd.concat([personal_data, price_data], axis=1)

# Display first few rows
print("\nMerged Data Preview:")
print(merged_data.head())

# Print shape of merged dataset
print("\nMerged Dataset Shape:", merged_data.shape)
````

````python
from sklearn.preprocessing import LabelEncoder

label_enc = LabelEncoder()
categorical_columns = ['sex', 'smoker', 'region']  

for col in categorical_columns:
    merged_data[col] = label_enc.fit_transform(merged_data[col])
````

````python
# Save the cleaned dataset
merged_data.to_csv("cleaned_insurance_data.csv", index=False)
````

````python
# Display final dataset info
print("\nFinal Cleaned Dataset Info:")
print(merged_data.info())

# Display summary statistics
print("\nSummary Statistics:")
print(merged_data.describe())
````

We first **checked for missing values** and handled them by filling numerical columns with the **median** and categorical columns with the **mode**.
